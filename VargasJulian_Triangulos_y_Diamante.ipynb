```python
# Actividad:
# Realice un programa con el lenguaje de Python donde solicite al usuario un digito.

# Este digito corresponde a las dimensiones de un triangulo, altura y base. El programa debe entregar cinco figuras,
# las primeras cuatro corresponde a un triangulo el cual rota 90º grados luego de la construcción de los triángulos
# estos deben generar un diamante.

n = int(input("Ingresa un numero: "))


print("\n")

for i in range(1,n+1):
  print("*" * i)

print("\n")


for i in range(n, 0, -1):
  print("*" * i)

print("\n")


for i in range(1, n+1):
  print(" " * (n - i) + "*" * i)

print("\n")


for i in range(n, 0, -1):
  print(" " * (n - i) + "*" * i)




# Parte Superior
for i in range(n + 1):
  print(" " * (n - i) + "*" * (2 * i))

# Parte Inferior
for i in range(n , 0, -1):
  print(" " * (n - i + 1) + "*" * (2 * i - 2))





```
