```python
import math


def identificarParInpar(num):
    if num % 2 == 0:
        print(f"El número {num} es par.")
    else:
        print(f"El número {num} es impar.")


def calculoFactorial(num):
    if num < 0:
        print("No se puede calcular el factorial de un número negativo.")
    elif num == 0:
        print("El factorial de 0 es 1.")
    else:
        factorial = 1
        for i in range(1, num + 1):
            factorial *= i
        print(f"El factorial de {num} es {factorial}.")


def conversionTemperatura(num, unidad):
    if unidad == 'c':
        fahrenheit = (num * 9/5) + 32
        print(f"{num}°C es igual a {fahrenheit}°F.")
    elif unidad == 'f':
        celsius = (num - 32) * 5/9
        print(f"{num}°F es igual a {celsius}°C.")
    else:
        print("Unidad no válida. Debe ser C o F.")


def mayorTresNumeros(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c


def calcularFiguras():
    num = int(input("'Calcular:\n 1 => cubo,\n 2 => esfera,\n 3 => triangulo,\n 4 => rectángulo,\n 5 => pirámide'\n"))

    if num == 1:
        lado = float(input("Ingrese el lado del cubo: "))
        area_superficial  = 6 * lado * lado
        volumen = lado * lado * lado
        print("El area superficial del cubo es: ", area_superficial)
        print("El volumen del cubo es: ", volumen)

    elif num == 2:
        radio = float(input("Ingrese el radio de la esfera: "))
        area_superficial = 4 * math.pi * radio * radio
        volumen = (4/3) * math.pi * radio * radio * radio
        print("El area superficial de la esfera es: ", area_superficial)
        print("El volumen de la esfera es: ", volumen)

    elif num == 3 :
        base = float(input("Ingrese la base del triángulo: "))
        altura = float(input("Ingrese la altura del triángulo: "))
        area = (base * altura) / 2
        perimetro = 3 * base
        print("El área del triángulo es: ", area)
        print("El perímetro del triángulo es: ", perimetro)

    elif num == 4 :
        base = float(input("Ingrese la base del rectángulo: "))
        altura = float(input("Ingrese la altura del rectángulo: "))
        area = base * altura
        perimetro = 2 * (base + altura)
        print("El área del rectángulo es: ", area)
        print("El perímetro del rectángulo es: ", perimetro)

    elif num == 5 :
        longitud_base = float(input("Ingrese la longitud de la base de la pirámide: "))
        numero_lados = float(input("Ingrese el número de lados de la base: "))
        altura = float(input("Ingrese la altura de la pirámide: "))
        apotema = longitud_base / (2 * math.tan(math.pi / numero_lados))
        area_base = (numero_lados * longitud_base * apotema) / 2

        altura_inclinada = math.sqrt(apotema**2 + altura**2)
        area_lateral = (numero_lados * longitud_base * altura_inclinada) / 2
        area_superficial = area_base + area_lateral
        volumen = (area_base * altura) / 3

        print("El área superficial de la pirámide es: ", area_superficial)
        print("El volumen de la pirámide es: ", volumen)

    else:
        "Selecciona una opcion valida"


1
while True:
    print("\n¿Que desea hacer?,\n\n 1 => Sumar dos números, \n 2 => Identificar Par o Inpar, \n 3 => Realizar un cálculo factorial")
    print(" 4 => Conversión Temperatura, \n 5 => Encontrar el mayor de 3 números,\n 6 => Calcular Valores de Figuras")
    print(" 7 => Cerrar \n")

    opcion = input("Opción (1-7): ")

    print("_________________________________________\n")

    if opcion == '1':
        print("1 => Suma de dos numeros")

        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
        print("La suma de los dos números es: ", a + b)

    elif opcion == '2':
        print("2 => Identificar par o inpar - Entero")
        num = int(input("Ingrese un número: "))

        identificarParInpar(num)


    elif opcion == '3':
        print("3 => Calculo factorial")
        try:
            num = int(input("Ingrese un número: "))

            calculoFactorial(num)

        except ValueError:
            print("Error: Debe ingresar un número entero.")

    elif opcion == '4':
        print("4 => Conversion temperatura")
        try:
            num = float(input("Ingrese la temperatura en valor numerico: "))
            unidad = input("Ingrese la unidad (C o F): ").lower()
            conversionTemperatura(num, unidad)
        except ValueError:
            print("Error: Debe ingresar un número entero.")

    elif opcion == '5':
        print("5 => Mayor de 3 numeros")
        try:
            a = float(input("Ingrese el primer número: "))
            b = float(input("Ingrese el segundo número: "))
            c = float(input("Ingrese el tercer número: "))
            print("El mayor de los tres números es: ", mayorTresNumeros(a, b, c))
        except ValueError:
            print("Error: Debe ingresar números enteros.")

    elif opcion == '6':
        print("6 => Calcular Valores de Figuras")
        calcularFiguras()

    elif opcion == '7':
        print("Hasta luego")
        break

    else:
        print(f"Opción \"{opcion}\" No es Válida")
```
