## Segundo ejercicio: Invirtiendo cadenas


```
/*
 * Crea un programa que invierta el orden de una cadena de texto
 * sin usar funciones propias del lenguaje que lo hagan de forma automática.
 * - Si le pasamos "Hola mundo" nos retornaría "odnum aloH"
 */
```



Análisis previo:

````python
# Invertir cadena de texto

def invertirCadena(cadena):

    listaCadena = list(cadena)

    caracteresN = len(listaCadena)
    invertida = ""
    for i in range(caracteresN - 1, -1, -1):
        invertida += listaCadena[i]
    return invertida

cadena = input("Ingrese su cadena: ")
print(invertirCadena(cadena))
````

## Cuarto ejercicio: ¿Es un palíndromo?
DJzpqQnWkAAq6xb.jpg

```
/*
 * Escribe una función que reciba un texto y retorne verdadero o
 * falso (Boolean) según sean o no palíndromos.
 * Un Palíndromo es una palabra o expresión que es igual si se lee
  * de izquierda a derecha que de derecha a izquierda.
 * NO se tienen en cuenta los espacios, signos de puntuación y tildes.
 * Ejemplo: Ana lleva al oso la avellana.
 */
```



Análisis previo:

````python
palindromo = input("Ingrese texto: ")

def confirmarPalidromo(palindromo):
  palindromo = palindromo.lower()
  palindromo = palindromo.replace(" ", "")
  palindromo = palindromo.replace(",", "")
  palindromo = palindromo.replace(".", "")

  listaPalin = list(palindromo)

  caracteresN = len(listaPalin)
  invertida = ""
  for i in range(caracteresN - 1, -1, -1):
      invertida += listaPalin[i]


  if palindromo == invertida:
    print("Palindromo: "+ palindromo)
    return True

  else:
    return False

print(confirmarPalidromo(palindromo))
````

## Sexto ejercicio: Cifrado Cesar


```
/*
 * Crea un programa que realize el cifrado César de un texto y lo imprima.
 * También debe ser capaz de descifrarlo cuando así se lo indiquemos.
 *
 * Te recomiendo que busques información para conocer en profundidad cómo
 * realizar el cifrado. Esto también forma parte del reto.
 */
```



Análisis previo:

````python
import ipywidgets as widgets
from IPython.display import display



# Cifrar o Descifrar
def cifrarDescifrar(texto, clave):
  alfabeto = "abcdefghijklmnopqrstuvwxyz"
  alfabetoUpper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

  resultado = ""
  for l in texto:

    if l in alfabeto:

      indice = alfabeto.index(l)
      indice += clave
      if indice >= len(alfabeto):
        indice -= len(alfabeto)
      if indice < 0:
        indice += len(alfabeto)

      resultado += alfabeto[indice]


    elif l in alfabetoUpper:

      indice = alfabetoUpper.index(l)
      indice += clave
      if indice >= len(alfabetoUpper):
        indice -= len(alfabetoUpper)
      if indice < 0:
        indice += len(alfabetoUpper)

      resultado += alfabetoUpper[indice]


    else:
      resultado += l
      # print("Resultado: ",resultado)

  return resultado

# Descifrar Varias Versiones
def descifrarAut(texto):

  alfabeto = "abcdefghijklmnopqrstuvwxyz"
  alfabetoUpper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

  for clave in range(len(alfabeto)):

    resultado = cifrarDescifrar(texto, -clave)
    print(f"Clave {-clave}: {resultado}")

# ipywidgets
textoWidget = widgets.Text(description="Ingrese texto:")

claveWidget = widgets.IntSlider(description="Ingrese clave:", min=1, max=25, value=1)

signoWidget = widgets.Dropdown(options=["+", "-"], description="Signo:")

salidaWidget = widgets.Output()

# Botones
botonCifrarDescifrar = widgets.Button(description="Cifrar/Descifrar")
botonDescifrarAut = widgets.Button(description="Descifrar Auto")

# Funciones de Botones
def cifrarDescifrarBoton(b):
  with salidaWidget:
    salidaWidget.clear_output()

    texto = textoWidget.value
    texto = texto.lower()
    texto = texto.replace(",", "")
    texto = texto.replace(".", "")

    magnitud = claveWidget.value
    signo = signoWidget.value
    clave = magnitud if signo == "+" else -magnitud
    resultado = cifrarDescifrar(texto, clave)
    print(f"Texto con clave {clave}: {resultado}")

def descifrarAutoBoton(b):
  with salidaWidget:
    salidaWidget.clear_output()
    texto = textoWidget.value
    descifrarAut(texto)

botonCifrarDescifrar.on_click(cifrarDescifrarBoton)

botonDescifrarAut.on_click(descifrarAutoBoton)

# Mostrar widgets
display(widgets.VBox([textoWidget,
                      claveWidget,
                      signoWidget,
                      botonCifrarDescifrar,
                      botonDescifrarAut,
                      salidaWidget]))
````
